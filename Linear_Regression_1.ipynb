```python
import pandas as pd
df=pd.read_csv("/content/car_purchasing.csv",encoding='latin1')
```

```python
df1=df.copy()
df1.describe()
```

```python
df1.info()
```

```python
#Drop categorical and unwanted coloumns
df1=df1.drop(columns=['customer name','customer e-mail','country','gender'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df1.corr(), annot=True, cmap='mako')
plt.show()
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
print(x)
```

```python
#Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
x=scaler.fit_transform(x)
y=scaler.fit_transform(y)
print(x)
```

```python
#Split data into training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1, shuffle=True,random_state=42)
```

Splitted whole 500 datas as

Test data:50(10%)

Training data:450(90%)

```python
#Train the model
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_model.score(x_train,y_train)
lr_pred=lr_model.predict(x_test)
```

```python
#Evaluation metrices
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
import numpy as np

# Predictions
y_train_pred = lr_model.predict(x_train)
y_test_pred = lr_model.predict(x_test)

# Explained Variance
ev = explained_variance_score(y_test, y_test_pred)
print("Explained Variance:", ev)

# R^2 Score
r2 = r2_score(y_test, y_test_pred)
print("R^2 Score:", r2)

# Mean Absolute Error
mae = mean_absolute_error(y_test, y_test_pred)
print("MAE:", mae)

# Mean Squared Error
mse = mean_squared_error(y_test, y_test_pred)
print("MSE:", mse)

# Root Mean Squared Error
rmse = np.sqrt(mse)
print("RMSE:", rmse)


```

```python
#Ridge regressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
import numpy as np

# Initialize models
ridge_model = Ridge(alpha=1.0)
lasso_model = Lasso(alpha=0.1)  # alpha can be tuned

# Fit models
ridge_model.fit(x_train, y_train)
lasso_model.fit(x_train, y_train)

# Predictions
y_pred_lr = lr_model.predict(x_test)
y_pred_ridge = ridge_model.predict(x_test)
y_pred_lasso = lasso_model.predict(x_test)

# Function to print evaluation metrics
def print_metrics(y_true, y_pred, model_name="Model"):
    print(f"--- {model_name} Metrics ---")
    print("Explained Variance:", explained_variance_score(y_true, y_pred))
    print("R^2 Score:", r2_score(y_true, y_pred))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("\n")

# Compare all models
print_metrics(y_test, y_pred_lr, model_name="Linear Regression")
print_metrics(y_test, y_pred_ridge, model_name="Ridge Regression")
print_metrics(y_test, y_pred_lasso, model_name="Lasso Regression")

```

```python
#Plot the comparisons
import matplotlib.pyplot as plt

# Predictions (already fitted in previous step)
y_pred_lr = lr_model.predict(x_test)
y_pred_ridge = ridge_model.predict(x_test)
y_pred_lasso = lasso_model.predict(x_test)

# Create scatter plot
plt.figure(figsize=(8,6))

plt.scatter(y_test, y_pred_lr, label='Linear Regression', alpha=0.7, color='blue')
plt.scatter(y_test, y_pred_ridge, label='Ridge Regression', alpha=0.7, color='green')
plt.scatter(y_test, y_pred_lasso, label='Lasso Regression', alpha=0.7, color='red')

# Plot ideal line (y = x)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)

# Labels and title
plt.xlabel("Actual Net Worth")
plt.ylabel("Predicted Net Worth")
plt.title("Linear vs Ridge vs Lasso Regression Predictions")
plt.legend()
plt.grid(True)
plt.show()
```
